```python
from rouge_score import rouge_scorer
from bert_score import score

# Define la función de evaluación
def evaluate_response(generated_response, reference_response):
    # Calcular ROUGE score
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    rouge_scores = scorer.score(reference_response, generated_response)

    # Calcular BERTScore
    P, R, F1 = score([generated_response], [reference_response], lang="es", verbose=True)

    # Devolver resultados
    return rouge_scores, P.mean().item(), R.mean().item(), F1.mean().item()

# Iterar a través de los 10 pares de preguntas y respuestas
for i in range(1, 11):
    # Obtener las variables dinámicamente usando globals()
    question_var = globals()[f'question{i}']
    reference_response_var = globals()[f'reference_response{i}']
    generated_response_var = globals()[f'generated_response{i}']

    # Evaluar la respuesta generada contra la respuesta de referencia
    rouge_scores, P, R, F1 = evaluate_response(generated_response_var, reference_response_var)

    # Imprimir las respuestas generadas y de referencia junto con las métricas
    print(f"Iteración {i}:")
    print(f"Question: {question_var}")
    print(f"Generated Response: {generated_response_var}")
    print(f"Reference Response: {reference_response_var}")
    print(f"ROUGE-1 score: {rouge_scores['rouge1']}")
    print(f"ROUGE-2 score: {rouge_scores['rouge2']}")
    print(f"ROUGE-L score: {rouge_scores['rougeL']}")
    print(f"BERTScore - Precision: {P}, Recall: {R}, F1: {F1}")
    print("-" * 50)
```
