```python
# Ruta local del archivo PDF
local_path = "/content/COCINA_II_MANUAL_Y_RECETARIO_PG_II_2019.pdf"

# Función para unir todas las páginas en un archivo .txt usando PyMuPDF
def unir_paginas_en_txt_pymupdf(file_path):
    try:
        # Abrir el archivo PDF
        pdf_document = fitz.open(file_path)

        # Ruta de salida para el archivo .txt combinado
        output_file = "/content/combined_text_gastronomia.txt"

        # Abrir el archivo de texto para escribir
        with open(output_file, "w", encoding="utf-8") as txt_file:
            # Recorrer todas las páginas del PDF
            for page_num in range(len(pdf_document)):
                page = pdf_document.load_page(page_num)
                page_text = page.get_text()

                # Escribir el contenido de la página en el archivo de texto
                txt_file.write(f"Página {page_num + 1}\n\n")
                txt_file.write(page_text)
                txt_file.write("\n\n---\n\n")  # Separador entre páginas

        print(f"Se ha creado el archivo combinado: {output_file}")
    except Exception as e:
        print(f"Ocurrió un error: {str(e)}")

# Llamar a la función para unir todas las páginas en un archivo .txt
unir_paginas_en_txt_pymupdf(local_path)
```

```python
import fitz  # PyMuPDF

# Ruta local del archivo PDF
local_path = "/content/COCINA_II_MANUAL_Y_RECETARIO_PG_II_2019.pdf"

# Función para extraer texto de un PDF usando PyMuPDF
def extract_text_from_pdf_pymupdf(pdf_path):
    try:
        # Abrir el archivo PDF
        pdf_document = fitz.open(pdf_path)
        text = ''

        # Recorrer todas las páginas del PDF
        for page_num in range(len(pdf_document)):
            page = pdf_document.load_page(page_num)
            page_text = page.get_text()

            # Agregar el contenido de la página al texto completo
            text += f"Página {page_num + 1}\n\n{page_text}\n\n---\n\n"  # Separador entre páginas

        return text
    except Exception as e:
        print(f"Ocurrió un error al extraer el texto: {str(e)}")
        return ''

# Función para formatear el texto en párrafos
def format_text_into_paragraphs(text):
    paragraphs = text.split('\n\n')
    formatted_text = '\n\n'.join(paragraph.strip() for paragraph in paragraphs if paragraph.strip())
    return formatted_text

# Función para guardar el texto en un archivo
def save_text_to_file(text, txt_path):
    try:
        with open(txt_path, 'w', encoding='utf-8') as file:
            file.write(text)
        print(f'Texto guardado en {txt_path}')
    except Exception as e:
        print(f"Ocurrió un error al guardar el texto: {str(e)}")

# Ruta de salida para el archivo .txt combinado
output_path = "/content/TextoFormateadoCocina.txt"

# Extraer texto del PDF usando PyMuPDF
pdf_text = extract_text_from_pdf_pymupdf(local_path)

# Formatear el texto en párrafos
formatted_text = format_text_into_paragraphs(pdf_text)

# Guardar el texto formateado en un archivo .txt
save_text_to_file(formatted_text, output_path)
```

```python
from mistralai.client import MistralClient
from mistralai.models.chat_completion import ChatMessage
import requests
import numpy as np
import faiss
import os
from getpass import getpass

api_key= getpass("Type your API Key")
client = MistralClient(api_key=api_key)
```

```python
text=formatted_text
```

```python
chunk_size = 2048
chunks = [text[i:i + chunk_size] for i in range(0, len(text), chunk_size)]
```

```python
def get_text_embedding(input):
    embeddings_batch_response = client.embeddings(
          model="mistral-embed",
          input=input
      )
    return embeddings_batch_response.data[0].embedding
```

```python
text_embeddings = np.array([get_text_embedding(chunk) for chunk in chunks])
```

```python
text_embeddings.shape
```

```python
text_embeddings
```

```python
d = text_embeddings.shape[1]
index = faiss.IndexFlatL2(d)
index.add(text_embeddings)
```

```python
question = "¿Qué técnicas existen para la confeccion de QUICHES: Masas secas, quebradas o friables?"
question_embeddings = np.array([get_text_embedding(question)])
question_embeddings.shape
```

```python
question_embeddings
```

```python
D, I = index.search(question_embeddings, k=2)
print(I)
```

```python
retrieved_chunk = [chunks[i] for i in I.tolist()[0]]
print(retrieved_chunk)
```

```python
prompt = f"""
La información de contexto está a continuación..
---------------------
{retrieved_chunk}
---------------------
Dada la información del contexto y no el conocimiento previo, responda la consulta.
Consulta: {question}
Respuesta:
"""
```

```python
def run_mistral(user_message, model="mistral-large-latest"):
    messages = [
        ChatMessage(role="user", content=user_message)
    ]
    chat_response = client.chat(
        model=model,
        messages=messages
    )
    return (chat_response.choices[0].message.content)
```

```python
# Ejecuta la función run_mistral con el prompt generado
response = run_mistral(prompt)

# La respuesta generada por el modelo se almacena en la variable 'response'
print(response)
```
